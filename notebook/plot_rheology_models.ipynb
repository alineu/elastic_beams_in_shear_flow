```python
import pylab
import matplotlib.pyplot as plt
import numpy

def get_cmap(n,name='viridis'):
    return plt.cm.get_cmap(name,n)

a = 2
nu0=1
nuinf=1e-3
h_channel = 1.2 #m
shear_rate_order = 10
x = numpy.logspace(-shear_rate_order,shear_rate_order,10000)
k=[3e-1,3e2,5e8]
n=[0.8]
fig, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
velocity = 2.0 #m/s
gap=0.4
shear_rate = velocity/gap #1/s
dim=len(k)*len(n)
cmap=get_cmap(dim)
right, bottom, width, height = [0.6, 0.6, 0.28, 0.25]
#ax2 = fig.add_axes([right, bottom, width, height])
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        ax.loglog(x,nu,':',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\mu$'+' (Pa.s)')
        ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
        ax.set_xlim([1e-9,1e9])        
ax.axvline(x=(shear_rate),c='r',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))
ax.set_title('Carreau-Yasuda model parameters \n ' + 
             r'$\mu=\mu_{\,\infty}+(\mu_{\,0}-\mu_{\,\infty})\big[{1+(k\dot{\gamma})^a}\big]^{\frac{{(n-1)}_{ }}{a}}$')
pylab.show()
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
gap=0.4
shear_rate = velocity/gap #1/s
#tau_yield=shear_rate*(nuinf+nu0)/2*numpy.ones(npoints)
#print(tau_yield[0])
dim=len(k)*len(n)
cmap=get_cmap(dim)
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        tau = np.multiply(nu,x)
        ax.loglog(x,tau,'-',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\tau$'+' (Pa)')
        ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
        ax.set_xlim([1e-7,1e8])
ax.axvline(x=(shear_rate),c='g',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))
pylab.show()





k=[0.3,5,1e5]
n=[0.5]
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
velocity = 2 #m/s
gap=0.6
shear_rate = velocity/gap #1/s
dim=len(k)*len(n)
cmap=get_cmap(dim)
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        ax.loglog(x,nu,':',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\mu$'+' (Pa.s)')
        ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
        ax.set_xlim([1e-6,1e6])
ax.axvline(x=(shear_rate),c='g',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))

pylab.show()
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
gap=0.6
shear_rate = velocity/gap #1/s
dim=len(k)*len(n)
cmap=get_cmap(dim)
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        tau = np.multiply(nu,x)
        ax.loglog(x,tau,'-',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\tau$'+' (Pa)')
        ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
        ax.set_xlim([1e-7,1e8])
ax.axvline(x=(shear_rate),c='g',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))
pylab.show()


k=[0.2,3,1e3]
n=[0.2]
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
velocity = 2 #m/s
gap=0.8
shear_rate = velocity/gap #1/s
dim=len(k)*len(n)
cmap=get_cmap(dim)
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        ax.loglog(x,nu,':',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\mu$'+' (Pa.s)')
        ax.set_xlabel(r'$\dot{\gamma} (\frac{1}{s})$')
        ax.set_xlim([1e-4,1e4])
ax.axvline(x=shear_rate,c='b',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))
pylab.show()
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
gap=0.8
shear_rate = velocity/gap #1/s
dim=len(k)*len(n)
cmap=get_cmap(dim)
for i_k in range(len(k)):
    for i_n in range(len(n)):
        index = i_k*len(n)+i_n
        nu = nuinf+(nu0-nuinf)*(1+(k[i_k]*x)**a)**((n[i_n]-1)/a)
        tau = np.multiply(nu,x)
        ax.loglog(x,tau,'-',c=cmap(int(index)),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (k[i_k],n[i_n]))
        ax.set_ylabel(r'$\tau$'+' (Pa)')
        ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
        ax.set_xlim([1e-5,1e5])
ax.axvline(x=(shear_rate),c='g',lw=0.5,linestyle='--',label=(r'$\dot{\gamma} = %2.2f $')%shear_rate)
ax.legend(loc='center',bbox_to_anchor=(1.2,0.5))
pylab.show()
```

```python
import itertools
n=[0.02,0.5]
k=[0.5,5]
taus=[1]

gaps = [0.6]
velocity = 2 #m/s
shear_rate_order = 9
min_shear_rate = 0.01
npoints = 500
cols=['r','g','b']
dim=(len(k)*len(n)*len(gaps)*len(taus))
cmap=get_cmap(dim)
x = numpy.logspace(-shear_rate_order,shear_rate_order,npoints)
points = numpy.ones(npoints)
f, ax = plt.subplots(figsize=(6,4),dpi=250,facecolor='w')
for i,bb in enumerate(itertools.product(k,n,gaps,taus)):
    kk,nn,gg,tt = bb
    tau_yield = tt
    shear_rate = velocity/gg
    #mu00 = tau_yield/min_shear_rate
    mu00 = 10
    #x0_tau0 = min_shear_rate
    #mu00=(tau_yield/min_shear_rate)+kk*(min_shear_rate**(nn-1))
    x0_tau0 = ((mu00-(tau_yield/shear_rate))/kk)**(1/(nn-1))
    x_tau0 = numpy.logspace(np.log10(x0_tau0/10000),np.log10(x0_tau0),100)
    
    ax.loglog(x_tau0,(tau_yield/shear_rate)+kk*(x_tau0**(nn-1)),':',c=cmap(i),mfc='w',markersize=2)#,label="k=%2.1e, n=%2.1e" % (kk,nn))
    #x = numpy.logspace(np.log10(x0_tau0),shear_rate_order,npoints)
    mu = numpy.minimum(mu00*points,(tau_yield/shear_rate)+kk*(x**(nn-1)))
    ax.loglog(x,mu,'-',c=cmap(i),mfc='w',markersize=2,label="k=%2.1e, n=%2.1e" % (kk,nn))
    #low_range = numpy.logspace(-shear_rate_order,np.log10(x0_tau0),npoints)
    #ax.loglog(low_range,mu00*points,':',c=cmap(i),mfc='w',markersize=2)
    ax.set_xlabel(r'$\dot{\gamma}  \ (\frac{1}{s})$')
    ax.set_ylim([1e-2,1e2])
    ax.set_ylabel(r'$\mu$'+' (Pa.s)')
    ax.set_xlim([1e-6,1e6])    
for i,gg in enumerate(gaps):
    shear_rate = velocity/gg
    ax.axvline(x=(shear_rate),c=cols[i],lw=0.5,linestyle='--',label=(r'$\mathrm{max}(\dot{\gamma}) = %2.2f $')%(shear_rate))

            
ax.set_title("Herschel-Bulkley model parameters \n " + 
             r'$\mu = \mathrm{min}(\mu_0,\frac{\tau_0}{\dot{\gamma}}+k\,\dot{\gamma}^{\,n-1})$')
ax.legend(loc='top',bbox_to_anchor=(0.6,1))
pylab.show()  

```
